```python
# member = df[['City', 'Customer type', 'gross income']].groupby(['Branch'])

print('===Gender====')

print("Female",df['Gender'].tolist().count('Female'))
print("Male",df['Gender'].tolist().count('Male'))

print('=====City==')

print("Yangon",df['City'].tolist().count('Yangon'))
print("Naypyitaw",df['City'].tolist().count('Naypyitaw'))
print("Mandalay",df['City'].tolist().count('Mandalay'))

# print('====Branch===')

# A - Yangyon
# B - Mandalay
# C - Naypyitaw


# print(df['Branch'].tolist().count('A'))
# print(df['Branch'].tolist().count('B'))
# print(df['Branch'].tolist().count('C'))

print('===Member Type====')

print("Member",df['Customer type'].tolist().count('Member'))
print("Normal",df['Customer type'].tolist().count('Normal'))


print('===Payment Type====')

# df['Payment'].unique()

print("Ewallet",df['Payment'].tolist().count('Ewallet'))
print("Cash",df['Payment'].tolist().count('Cash'))
print("Credit card",df['Payment'].tolist().count('Credit card'))



print('===Payment Type====')

df['Product line'].unique()

print("Health and beauty",df['Product line'].tolist().count('Health and beauty'))
print("Electronic accessories",df['Product line'].tolist().count('Electronic accessories'))
print("Home and lifestyle",df['Product line'].tolist().count('Home and lifestyle'))
print("Sports and travel",df['Product line'].tolist().count('Sports and travel'))
print("Food and beverages",df['Product line'].tolist().count('Food and beverages'))
print("Fashion accessories",df['Product line'].tolist().count('Fashion accessories'))

display(df.groupby(['Product line','Gender',]).size())


```

```python
# pd.pivot_table(df,index='Invoice ID')

pivot_df = (pd.pivot_table(df, index = ['Payment']
            ,values=['cogs']
            ,aggfunc={'cogs':np.sum}
                          ))
display(pivot_df)

pivot_df = (pd.pivot_table(df, index = ['City']
            ,values=['cogs']
            ,aggfunc={'cogs':np.sum}
                          ))
display(pivot_df)

pivot_df = (pd.pivot_table(df, index = ['Product line']
            ,values=['cogs']
            ,aggfunc={'cogs':np.sum}
                          ))
display(pivot_df)




pivot_df = (pd.pivot_table(df, index = ['Product line']
            ,values=['cogs']
            ,aggfunc={'cogs':np.average}
                          ))
display(pivot_df)


pivot_df = (pd.pivot_table(df, index = ['Product line']
            ,values=['gross income']
            ,aggfunc={'gross income':np.sum}
                          ))
display(pivot_df)



pivot_df = (pd.pivot_table(df, index = ['Product line', 'City']
            ,values=['cogs']
            ,aggfunc={'cogs':np.sum}
                          ))
display(pivot_df)

pivot_df = (pd.pivot_table(df, index = ['Gender', 'Product line']
            ,values=['cogs']
            ,aggfunc={'cogs':np.sum}
                          ))
display(pivot_df)

pivot_df = (pd.pivot_table(df, index = ['Customer type']
            ,values=['cogs']
            ,aggfunc={'cogs':np.sum}
                          ))
display(pivot_df)



# ,
#                           'Quantity':[np.min,np.max]
```
